```python
import pandas as pd
import psycopg2
import configparser
```

# postgres to local

```python
config = configparser.ConfigParser()
config.read('../../../config.ini')

user = config['postgresql']['user']
password = config['postgresql']['password']
host = config['postgresql']['host']
database = config['postgresql']['database']

connection_string = f"postgresql://{user}:{password}@{host}/{database}"
```

```python
# PostgreSQLに接続
%load_ext sql
%sql $connection_string
```

```python
%%sql
SELECT *
FROM jvd_se
LIMIT 5;
```

```python
# SQLクエリを実行し、指定したカラムを取得
uma_race_sql = %sql SELECT * FROM jvd_se WHERE CAST(kaisai_nen AS INTEGER) >= 2010;

# 結果をDataFrameに変換
uma_race_df = pd.DataFrame(uma_race_sql)
uma_race_df.info()
```
