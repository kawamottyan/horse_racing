```python
import pandas as pd
import psycopg2
import configparser
```

# postgres to local

```python
config = configparser.ConfigParser()
config.read('../../../config.ini')

user = config['postgresql']['user']
password = config['postgresql']['password']
host = config['postgresql']['host']
database = config['postgresql']['database']

connection_string = f"postgresql://{user}:{password}@{host}/{database}"
```

```python
# PostgreSQLに接続
%load_ext sql
%sql $connection_string
```

```python
%%sql
SELECT *
FROM jvd_hr
LIMIT 5;
```

```python
%%sql
SELECT *
FROM nvd_hr
LIMIT 5;
```

```python
# SQLクエリを実行し、指定したカラムを取得
payout_sql = %sql SELECT * FROM jvd_hr WHERE CAST(kaisai_nen AS INTEGER) >= 2010;
n_payout_sql = %sql SELECT * FROM nvd_hr WHERE CAST(kaisai_nen AS INTEGER) >= 2010;

# 結果をDataFrameに変換
payout_df = pd.DataFrame(payout_sql)
n_payout_df = pd.DataFrame(n_payout_sql)
payout_df.info()
```

# save as pickle

```python
payout_df.to_pickle('../../traindata/rawdata/payout.pkl')
n_payout_df.to_pickle('../../traindata/rawdata/n_payout.pkl')
```
