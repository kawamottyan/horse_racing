```python
import pandas as pd
import psycopg2
import configparser
import datetime
```

# postgres to local

```python
config = configparser.ConfigParser()
config.read('../../../config.ini')

user = config['postgresql']['user']
password = config['postgresql']['password']
host = config['postgresql']['host']
database = config['postgresql']['database']

connection_string = f"postgresql://{user}:{password}@{host}/{database}"
```

```python
# PostgreSQLに接続
%load_ext sql
%sql $connection_string
```

```python
%%sql
SELECT *
FROM nvd_se
LIMIT 5;
```

```python
# 今日の日付を取得
today = datetime.date.today()
# 年と月日の形式に分解
year = today.year
month_day = today.strftime('%m%d')
```

```python
# SQLクエリの作成
n_uma_race_query = f"""
SELECT * 
FROM nvd_se 
WHERE (CAST(kaisai_nen AS INTEGER) > {year} OR (CAST(kaisai_nen AS INTEGER) = {year} AND CAST(kaisai_tsukihi AS INTEGER) >= {month_day}));
"""
```

```python
# SQLクエリの作成
uma_race_query = f"""
SELECT * 
FROM apd_sokuho_se 
WHERE (CAST(kaisai_nen AS INTEGER) > {year} OR (CAST(kaisai_nen AS INTEGER) = {year} AND CAST(kaisai_tsukihi AS INTEGER) >= {month_day}));
"""

uma_race_sql = %sql $uma_race_query
```

```python
# SQLクエリの作成
n_race_query = f"""
SELECT * 
FROM apd_sokuho_ra
WHERE (CAST(kaisai_nen AS INTEGER) > {year} OR (CAST(kaisai_nen AS INTEGER) = {year} AND CAST(kaisai_tsukihi AS INTEGER) >= {month_day}));
"""
```

```python
# SQLクエリを実行し、指定したカラムを取得
n_uma_race_sql = %sql $n_uma_race_query

# 結果をDataFrameに変換
n_uma_race_df = pd.DataFrame(n_uma_race_sql)
```
