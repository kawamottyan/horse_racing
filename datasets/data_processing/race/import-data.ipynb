# preparation

```python
import pandas as pd
import psycopg2
import configparser
```

# postgres to local

```python
config = configparser.ConfigParser()
config.read('../../../config.ini')

user = config['postgresql']['user']
password = config['postgresql']['password']
host = config['postgresql']['host']
database = config['postgresql']['database']

connection_string = f"postgresql://{user}:{password}@{host}/{database}"
```

```python
# PostgreSQLに接続
%load_ext sql
%sql $connection_string
```

```python
%%sql
SELECT *
FROM jvd_ra
LIMIT 5;
```

```python
# SQLクエリを実行し、指定したカラムを取得
j_race_sql = %sql SELECT * FROM jvd_ra WHERE CAST(kaisai_nen AS INTEGER) >= 2010;
n_race_sql = %sql SELECT * FROM nvd_ra WHERE CAST(kaisai_nen AS INTEGER) >= 2010;

# 結果をDataFrameに変換
j_race_df = pd.DataFrame(j_race_sql)
n_race_df = pd.DataFrame(n_race_sql)
j_race_df.info()
```

# save as pickle

```python
j_race_df.to_pickle('../../traindata/rawdata/j_race.pkl')
n_race_df.to_pickle('../../traindata/rawdata/n_race.pkl')
```
