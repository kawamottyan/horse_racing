```python
import pandas as pd
import psycopg2
import configparser
```

# postgres to local

```python
config = configparser.ConfigParser()
config.read('../../../config.ini')

user = config['postgresql']['user']
password = config['postgresql']['password']
host = config['postgresql']['host']
database = config['postgresql']['database']

connection_string = f"postgresql://{user}:{password}@{host}/{database}"
```

```python
# PostgreSQLに接続
%load_ext sql
%sql $connection_string
```

```python
%%sql
SELECT *
FROM n_race
LIMIT 5;
```

```python
# SQLクエリを実行し、指定したカラムを取得
race_sql = %sql SELECT * FROM n_race;

# 結果をDataFrameに変換
race_df = pd.DataFrame(race_sql)
race_df.info()
```

# save as pickle

```python
race_df.to_pickle('../../rawdata/race.pkl')
```
